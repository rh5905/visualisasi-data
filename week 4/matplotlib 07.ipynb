# Sample Data

```python
data_uas=[
    ['Bejo', 70],
    ['Tejo', 83],
    ['Cecep', 62],
    ['Wati', 74],
    ['Karti', 71]
]
```

```python
data_uas
```

# Table Plot dengan OO Style

```python
fig, ax = plt.subplots()

table = plt.table(cellText=data_uas, 
                  loc='center')

table.set_fontsize(14)
table.scale(1, 4) # ukuran kolom, baris

ax.axis(False)

plt.show()
```
